```python
continuous_columns = heart_data.drop(columns=['sex', 'cp', 'fbs', 'exang', 'target'])
categorical_columns = heart_data[['sex', 'cp', 'fbs', 'exang', 'target']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standardized_data = scaler.fit_transform(continuous_columns)
heart_data = pd.concat([pd.DataFrame(standardized_data, columns=continuous_columns.columns),categorical_columns], axis= 1)
```

```python
heart_data.shape
```

```python
X = heart_data.drop(columns='target', axis=1)
y = heart_data['target']
```

```python
print(X, y)
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
models = [LogisticRegression(), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_train_test():
    for model in models:
        # Training the model
        model.fit(X_train, y_train)

        # Evaluating the model
        test_pred = model.predict(X_test)
        test_acc = accuracy_score(test_pred, y_test)
        print(f'Accuracy score of the model {model}: {test_acc}')
```

```python
compare_models_train_test()
```

# Cross Validation

```python
cv_score_lr = cross_val_score(LogisticRegression(), X, y, cv= 5)
```

```python
print(cv_score_lr.mean())
```

```python
def compare_models_cv():
    for model in models:
        # Evaluating Cross Validation score
        cv_score = cross_val_score(model, X, y, cv= 5)

        print('-'*8, model, '-'*8)
        print(f'Cross Validation accuracies: {cv_score}')
        print(f'Accuracy % : {cv_score.mean().round(3) * 100}%')
```

```python
compare_models_cv()
```
