```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
raw_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(raw_cancer_data)
```

```python
cancer_data = pd.DataFrame(raw_cancer_data.data, columns=raw_cancer_data.feature_names)
```

```python
cancer_data['label'] = raw_cancer_data.target
```

```python
cancer_data.head()
```

```python
cancer_data.isnull().sum().sum()
```

```python
cancer_data.shape
```

```python
cancer_data.describe()
```

```python
cancer_data['label'].value_counts()
```

1 --> Belign

0 --> Malignant

```python
X = cancer_data.drop('label', axis=1)
y = cancer_data['label']
```

```python
print(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#scaler = StandardScaler()
#X_train = scaler.fit_transform(X_train)
#X_test = scaler.transform(X_test)
```

```python
model = SVC()
```

# Grid Search

```python
# Hyperparameter tuning
parameters = {
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'C' : [1, 5, 10, 20]
}
```

```python
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.cv_results_
```

```python
best_params = classifier.best_params_
best_params
```

```python
print(classifier.best_score_)
```

```python
# Loading results to DataFrame

result = pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
result.sort_values(by='mean_test_score', ascending=False)[['params', 'mean_test_score']].reset_index(drop=True)
```

```python
grid_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
grid_search_result
```

```python
best_model = classifier.best_estimator_
```

```python
best_model.score(X_test, y_test)
```

# **Randomized Search**

```python
model = SVC()
```

```python
# Hyperparameter tuning
parameters = {
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'],
    'C' : [1, 5, 10, 20]
}

classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.cv_results_
```

```python
best_params = classifier.best_params_
best_params
```

```python
print(classifier.best_score_)
```

```python
# Loading results to DataFrame

result = pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
result.sort_values(by='mean_test_score', ascending=False)[['params', 'mean_test_score']].reset_index(drop=True)
```

```python
randomized_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
randomized_search_result
```

```python
best_model = classifier.best_estimator_
```

```python
best_model.score(X_test, y_test)
```
