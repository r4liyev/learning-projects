# Pillow

```python
from PIL import Image
```

```python
img = Image.open(r"C:\Users\user\Downloads\puppy-facts.jpg")
img_resized = img.resize((200, 200))
```

```python
img_resized.size
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
img_resize = mpimg.imread('dog_image_resized.jpg')
img_plot = plt.imshow(img_resize)
plt.show()
```

# OpenCV (cv2)

```python
import cv2
```

```python
img = cv2.imread(r'C:\Users\user\Downloads\puppy-facts.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscale_img.shape
```

```python
plt.imshow(grayscale_img, cmap='gray')
```

```python
from IPython.display import display

pil_image = Image.fromarray(grayscale_img)
display(pil_image)
```

```python
cv2.imwrite('dog_grayscale_img.jpg', grayscale_img)
```
